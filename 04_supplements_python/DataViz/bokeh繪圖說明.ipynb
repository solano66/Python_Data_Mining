邊框設置

```python
p = figure(x_axis_label='index', 
           y_axis_label='price')

p.line(x, y, line_width=3)

p.diamond(x, y, fill_color="white", size = 10)

p.outline_line_width = 3
p.outline_line_alpha = 0.7
p.outline_line_color = "#ff3399"

show(p)
```

```python
p = figure(plot_width=400, plot_height=400)

x = diamonds.carat.head(30)
y = diamonds.price.head(30)

r = p.circle(x, y)

r.glyph.size = 10

r.glyph.fill_alpha = 0.5
r.glyph.fill_color = 'yellow'

r.glyph.line_color = "firebrick"
r.glyph.line_dash = [5, 1]
r.glyph.line_width = 2


show(p)
```

```python
p = figure(plot_width=400, plot_height=400, tools="tap")

p.circle(x, y,
         size = 20,
         selection_fill_color="yellow",
         selection_line_color="purple",
         selection_line_alpha=0.2,
         
         nonselection_fill_alpha=0.2,
         nonselection_fill_color="grey",
         nonselection_line_color="firebrick",
         nonselection_line_alpha=1.0
        )


show(p)
```

```python
from bokeh.sampledata.autompg import autompg
from bokeh.models import LinearColorMapper, ColorBar, ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256

source = ColumnDataSource(diamonds)

color_mapper = LinearColorMapper(palette=Viridis256, low=diamonds.depth.min(), high=diamonds.price.max())

p = figure(x_axis_label='Carat', y_axis_label='Depth', tools='', toolbar_location='above')

p.circle(x='carat', y='depth', 
         color={'field': 'price', 'transform': color_mapper}, 
         size=20, alpha=0.6, source=source)

color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0,0), title='Weight')

p.add_layout(color_bar, 'right')

show(p)
```

```python
date = pd.date_range('2018-07-01', '2018-08-31')
ts = pd.DataFrame({'Date' : date , 
                   'Temperature': np.random.randint(low = 20, 
                                                    high = 40, 
                                                    size=len(date))
                  })
x = ts.Date
y = ts.Temperature


p = figure(width=600, height=300, x_axis_type="datetime",  y_range=(0, 45))

p.line(x, y, line_dash="4 4", line_width=1, color='gray')

cr = p.circle(x, y, size=20,
              fill_color="grey", hover_fill_color="firebrick",
              fill_alpha=0.05, hover_alpha=0.3,
              line_color=None, hover_line_color="white")

p.add_tools(HoverTool(tooltips=None, renderers=[cr], mode='hline'))

show(p)
```

設定軸樣式

```python
p.yaxis.axis_label = "Temperature"
p.xaxis.axis_label = "Date"

p.axis.major_label_text_color = "brown"

show(p)
```

```python
p.xaxis.major_label_orientation = 45
p.yaxis.major_label_orientation = "vertical"

show(p)
```

```python
p.xaxis.axis_line_width = 2

p.yaxis.axis_line_color = "blue"

p.yaxis.major_label_text_color = "green"
p.yaxis.major_label_orientation = "vertical"

p.axis.minor_tick_in = -2
p.axis.minor_tick_out = 1

show(p)
```

```python
p.ygrid.grid_line_color = None

p.xgrid.grid_line_alpha = 0.7
p.xgrid.grid_line_dash = [2, 3]
p.xgrid.grid_line_color = '#2D406C'
p.xgrid.grid_line_width = 2

show(p)
```

```python
p.ygrid.band_fill_alpha = 0.1
p.ygrid.band_fill_color = "navy"

show(p)
```

添加標籤


```python
from bokeh.models.annotations import Label

temp = ts.Temperature
high_temp = temp.max()
idx = x[temp == high_temp].tolist()[0]

text = 'High Temp : {}'.format(high_temp)

label = Label(x=idx, y=high_temp, x_offset=12, text=text, text_baseline="middle")

p.add_layout(label)

show(p)
```

添加水平/垂直線

```python
from bokeh.models.annotations import Span

p = figure(plot_width=400, plot_height=400)

y = diamonds.depth.head(50)
x = range(len(y))

p.line(x, y)

h = y.mean()
v = 25

hline = Span(location=h, dimension='width', line_color='olive', line_width=3)
p.add_layout(hline)

vline = Span(location=v, dimension='height', line_color='olive', line_width=3)
p.add_layout(vline)

show(p)
```

設定佈局

```python
from bokeh.layouts import row, column, gridplot

p1 = figure(x_axis_label='carat', 
            y_axis_label='price',
            plot_width=200, 
            plot_height=200)

x = diamonds.carat
y = diamonds.price
p1.circle(x, y, size=3)

p2 = figure(x_axis_label='depth', 
            y_axis_label='price',
            plot_width=200, 
            plot_height=200)

x = diamonds.depth
y = diamonds.price
p2.circle(x, y, size=3)

p3 = figure(x_axis_label='table', 
            y_axis_label='price',
            plot_width=200, 
            plot_height=200)

x = diamonds.table
y = diamonds.price
p3.circle(x, y, size=3)


show(row(p1, p2, p3))
```

```python
show(column(p1, p2, p3))
```

```python
p = gridplot([[p1, None, None], [None, p2, None,], [None, None, p3]], toolbar_location=None)

show(p)
```

圖形連動

```python
x = diamonds.price.tolist()
y0 = diamonds.carat.tolist()
y1 = diamonds.depth.tolist()

# create a column data source for the plots to share
source = ColumnDataSource(data=dict(x=x, y0=y0, y1=y1))

TOOLS = "box_select,lasso_select,help"

# create a new plot and add a renderer
left = figure(tools=TOOLS, width=300, height=300)
left.circle('x', 'y0', source=source)

# create another new plot and add a renderer
right = figure(tools=TOOLS, width=300, height=300)
right.circle('x', 'y1', source=source)

p = gridplot([[left, right]])

show(p)
```
