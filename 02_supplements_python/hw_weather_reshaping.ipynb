## 作業一：用Python將長寬混雜的weathers.txt清理完成

> 首先讀入weathers.txt資料集

```python
import pandas as pd
weathers = pd.read_table("./data/weather.txt")
```

### 所謂長中帶寬的意思是年、月、最高和最低溫是長格式，而日期為寬格式，！！！

### 所謂寬中帶長的意思是日期為寬格式，而年、月、最高和最低溫是長格式！！！

```python
weathers.head()
```

> 接著將寬表d1:d31轉成長表

```python
weathers_f1 = pd.melt(weathers, id_vars=['id', 'year', 'month','element'], var_name='day', value_name='value')
weathers_f1.head()
```

> 清理欄位day的文數字混雜的狀況

```python
weathers_f1['day'] = weathers_f1['day'].str.extract('(\d+)').astype(int) # Check it in regexp101
weathers_f1.head()
```

> 將表依年月日升冪排序

```python
weathers_f2 = weathers_f1.sort_values(by=['year', 'month', 'day'])
weathers_f2.head()
```

> 將長表部份element下的TMAX和TMIN，轉成寬格式

- 與一開始讀入之表剛好反過來：Day從寬轉長了，而element從長轉成寬格式

```python
weathers_f3 = weathers_f2.pivot_table(index=['id', 'year','month','day'], columns=['element'], values='value') # 自動排除nan
weathers_f3
```

> 轉回原讀入形式

```python
weathers_b = weathers_f2.pivot_table(index=['id', 'year','month', 'element'], columns=['day'], values='value').reset_index()
weathers_b.head()
```
