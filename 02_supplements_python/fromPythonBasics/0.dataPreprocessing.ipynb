## 群組與摘要

> 資料匯入與理解

- phone_data.csv有數個月的行動電話使用記錄， 載入檔案檢視數據維度與維數、變數型別、與前五筆觀測值，其中型別為object的變數表示內容是字串。

```python
# 載入必要套件
import pandas as pd
import numpy as np
import dateutil
# 載入csv檔
path = './data'
fname = '/phone_data.csv'
data = pd.read_csv(''.join([path, fname]))
# 830筆觀測值，7個變數
print(data.shape)
```

```python
# 除index與duration外，所有欄位都是字串型別的類別變數
print(data.dtypes)
```

```python
print(data.head())
```

- 大數據時代下，資料多帶有時間戳記(time stamp)的脈絡(contextual)變數，將這些原為字串變數的時間，轉為跨時區可計算的時間格式變數，是處理時間序列資料的重要任務。
- 運用**pandas**序列資料結構的apply()方法，將序列`data['date']`中的每個字串元素轉換為`datetime64[ns]`時間格式，`ns`表示時間單位為`nanosecond`(\url{https://docs.scipy.org/doc/numpy-1.13.0/reference/arrays.datetime.html})。

```python
# 將日期字串逐一轉為時間格式
data['date'] = data['date'].apply(dateutil.parser.parse,
dayfirst=True)
# 也可以運用pandas的to_datetime()方法
data['date'] = pd.to_datetime(data['date'])
```

```python
# 'date'的資料型別已改變
print(data.dtypes)
```

- 前面**pandas**序列資料結構的`apply()`方法，與R語言的`apply()`函數一樣，都是隱式迴圈，其將內建函數、匿名函數或自定義函數等施加於序列中的每個元素，以下以一個短序列簡例說明`apply()`用法。

```python
# 傳入原生串列物件創建pandas序列，index引數給定橫向索引
series = pd.Series([20, 21, 12], index=['London',
'New York','Helsinki'])
print(series)
```

```python
# pandas序列物件apply()方法的多種用法
# 可套用內建函數，例如：對數函數np.log()
print(series.apply(np.log))
```

```python
# 也可以套用關鍵字為lambda的匿名函數
# 其x代表序列物件的各個元素
print(series.apply(lambda x: x**2))
```

```python
# 或是自定義函數square()
def square(x):
    return x**2
print(series.apply(square))
```

```python
# 另一個自定義函數，請注意參數custom_value如何傳入
def subtract_custom_value(x, custom_value):
    return x - custom_value
# 以args引數傳入值組(5,)作為custom_value參數
print(series.apply(subtract_custom_value, args=(5,)))
```

```python
# 各電信營運商語音通話總和
print(data[data['item'] == 'call'].groupby('network')
['duration'].sum())
```

- 群組與摘要也可以依多個欄位將數據分組後再進行統計。

```python
# 多個欄位分組，各月各服務類型的資料筆數統計
# 抓出分組數據的任何欄位統計筆數均可，此處以date為例
print(data.groupby(['month', 'item'])['date'].count())
```

- 資料處理與分析的工作有許多魔鬼藏在細節中，例如：分組結果可以**pandas**序列或資料框兩種方式呈現，其語法區別在於是否多了一對中括弧，細心的讀者可由下面的結果觀察到此一差異。

```python
# 分組統計結果pandas序列(duration變數名稱在最下面)
print(data.groupby('month')['duration'].sum())
print(type(data.groupby('month')['duration'].sum()))
```

```python
# 分組統計結果pandas資料框(duration變數名稱在上方)
print(data.groupby('month')[['duration']].sum())
print(type(data.groupby('month')[['duration']].sum()))
```

- 除了上面的`sum()`與`count()`統計方式，分組後的數據可以`agg()`方法決定統計的欄位與計算方法，欄位與計算是以兩者形成的鍵值原生字典結構傳入`agg()`函數中。
- 傳出的結果預設是以分組變數為索引的統計表，使用者可在數據分組時以`as_index=False`改變預設設定，傳回分組變數與欲統計變數的兩欄資料框。

```python
# 群組數據後的agg()分組統計
print(data.groupby('month').agg({"duration": "sum"})) # 結果與前面相同！
```

```python
# 分組索引值為分組變數month的值
print(data.groupby('month').agg({"duration": "sum"}).index)
```

```python
# 分組統計的欄位名稱為duration
print(data.groupby('month').agg({"duration": "sum"}).columns)
```

```python
# as_index=False改變預設設定，month從索引變成變數
print(data.groupby('month', as_index=False).agg({"duration":
"sum"}))
```

- 最後，`agg()`方法如果傳入多個鍵值對(key-value pairs)，即可對分組數據做如下的多個統計計算。

```python
# 各組多個統計計算
# 各月(month)各服務(item)的服務時間、網路服務型式與日期統計
print(data.groupby(['month', 'item']).agg({'duration':
[min, max, sum], 'network_type': "count", 'date':
[min, 'first', 'nunique']}))
```

> Box-Cox公式(只接受一維陣列，自動估計lambda)

\begin{equation}
x^{'} = \left\{\begin{matrix}
\frac{x^{\lambda}-1}{\lambda}, \textrm{if} \: \lambda \neq 0 \\ \log x, \textrm{if} \: \lambda=0.
\end{matrix}\right.
\end{equation}

- 根據量化屬性間的相關係數過濾贅餘的數值屬性(32個)

```python
def find_correlation(df, thresh=0.9):
    """
    Given a numeric pd.DataFrame, this will find highly correlated features,
    and return a list of features to remove
    params:
    - df : pd.DataFrame
    - thresh : correlation threshold, will remove one of pairs of features with a correlation greater than this value
    """
    
    corrMatrix = df.corr()
    corrMatrix.loc[:,:] =  np.tril(corrMatrix, k=-1) # 取下三角矩陣

    already_in = set() # 集合結構避免重複計入相同元素
    result = []

    for col in corrMatrix:
        perfect_corr = corrMatrix[col][corrMatrix[col] > thresh].index.tolist() # Index物件轉為list
        if perfect_corr and col not in already_in:
            already_in.update(set(perfect_corr)) # 更新集合
            perfect_corr.append(col)
            result.append(perfect_corr)
            
    select_nested = [f[1:] for f in result]
    select_flat = [i for j in select_nested for i in j]
    return select_flat
```

> 小結：資料前處理

- 資料前處理技術泛指訓練集資料的新增、刪除與轉換，預測建模前的資料準備對模型的預測能力有很大的影響。
- 減低資料偏斜或離群值對預測能力的影響，會大幅提升模型的績效。
- 較簡單的前處理策略是移除缺乏訊息的預測變數。
- 選用的分析模型類型決定資料前處理的需求，例如：樹狀模型對於預測變數的特性不敏感，而線性迴歸就並非如此了。
- 如何編碼預測變數(屬性工程)對模型績效有很大的影響，例如：兩變數的比值可能比各別變數還有用，但最有效的數據編碼可能來自於領域知識，而非數學技術。
- 同一預測變數有許多不同的編碼方式：
  - 距離一參考日期的天數，
  - 將年、月、日與週幾分成不同的預測變數，
  - 編碼成一年的第幾天，
  - 將日期編碼成上課日、假日或寒暑假。
- 總結來說，合宜的屬性工程取決於下列兩個因素：選用的模型、以及預測變數與結果變數之間的關係。
