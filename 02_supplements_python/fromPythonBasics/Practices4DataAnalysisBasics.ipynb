```python
df[4:10] # 其實是4到9，就是第5個到第10個
```

- You can also apply the standard row/column notation, by using the method iloc.

> The differences netween numpy arrays and pandas DataFrame:

- numpy handles “rows” first. E.g., data[0] is the first row of an array
- pandas starts with the columns. E.g., df['values'][0] is the first element of the column 'values'.
- If a DataFrame has labeled rows, you can extract for example the row “rowlabel” with df.loc['rowlabel']

```python
df[0:5]
```

- A sometimes confusing convention is that if you want to slice out a single row, e.g., row “5,” you have to use df[5:6]. If you use df[5] alone, you get an error!

```python
# df[5]
df[5:6]
```
